```python
import pandas as pd

df = pd.read_csv('Fake_real_Data.csv')
df.head()
```

```python
df.shape
```

```python
df.label.value_counts()
```

```python
df['label_num'] = df['label'].map({'Fake': 0, 'Real': 1})
df.head()
```

```python
import spacy
nlp = spacy.load('en_core_web_lg')
```

```python
#vectorizing a single sentence
doc = nlp('Top Trump Surrogate BRUTALLY Stabs Him')
doc.vector.shape
```

```python
df['vector'] = df['Text'].apply(lambda x: nlp(x).vector)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df.vector.values,
    df.label_num,
    test_size = 0.2,
    random_state = 1,
)
```

```python
X_train
```

```python
X_train.shape
```

```python
X_train
```

```python
#converting into 2-D array for training(must)

import numpy as np
X_train_2d = np.stack(X_train)
X_test_2d = np.stack(X_test)
```

```python
X_train_2d
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler #to tackle the negative values of the array

scaler = MinMaxScaler()
scaled_train_embed = scaler.fit_transform(X_train_2d)
scaled_test_embed = scaler.transform(X_test_2d)

clf = MultinomialNB()
clf.fit(scaled_train_embed, y_train)
```

```python
from sklearn.metrics import classification_report
y_pred = clf.predict(scaled_test_embed)

print(classification_report(y_test, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier(n_neighbors = 5, metric = 'euclidean')

clf.fit(X_train_2d, y_train)

y_pred = clf.predict(X_test_2d)

print(classification_report(y_test, y_pred))
```
